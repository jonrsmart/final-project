```python
daily_crypto = pd.read_csv('data/crypto_daily_prices.csv')
daily_crypto['date'] =  pd.to_datetime(daily_crypto['date'])
daily_crypto.head()
```

```python
daily_crypto.columns
```

```python
us_crypto = daily_crypto.drop(['1a. open (EUR)', '2a. high (EUR)', '3a. low (EUR)','4a. close (EUR)'], axis=1)
us_crypto.columns = ['Date','Open','High','Low','Close','Volume','Market Cap','Coin']
us_crypto.head()
```

```python
us_crypto['Date'] =  pd.to_datetime(us_crypto['Date'])
```

```python
us_crypto['200_MA'] = us_crypto['Close'].rolling(window=200).mean()
us_crypto['150_MA'] = us_crypto['Close'].rolling(window=150).mean()
us_crypto['50_MA'] = us_crypto['Close'].rolling(window=50).mean()
us_crypto['Volatility']=(us_crypto['High']-us_crypto['Low'])/us_crypto['Open']
us_crypto.set_index(['Coin'], inplace=True)
us_crypto.head()
```

```python
btc_df = us_crypto.loc['BTC']
eth_df = us_crypto.loc['ETH']
doge_df = us_crypto.loc['DOGE']
doge_df.head()
```

```python
doge_df.reset_index(inplace=True)
doge_df.set_index('Date', inplace=True)
```

```python
btc_df.to_csv('data/btc_data.csv')
eth_df.to_csv('data/eth_data.csv')
doge_df.to_csv('data/doge_data.csv')
```
