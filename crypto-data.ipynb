```python
daily_crypto = pd.read_csv('data/crypto_daily_prices.csv')
daily_crypto['date'] =  pd.to_datetime(daily_crypto['date'])
daily_crypto.head()
```

```python
daily_crypto.columns
```

```python
us_crypto = daily_crypto.drop(['1a. open (EUR)', '2a. high (EUR)', '3a. low (EUR)','4a. close (EUR)'], axis=1)
us_crypto.columns = ['Date','Open','High','Low','Close','Volume','Market Cap','Coin']
us_crypto.head()
```

```python
us_crypto['Date'] =  pd.to_datetime(us_crypto['Date'])
us_crypto.set_index(['Coin'], inplace=True)
us_crypto.head()
```

```python
btc_df = us_crypto.loc['BTC']
eth_df = us_crypto.loc['ETH']
doge_df = us_crypto.loc['DOGE']
doge_df.head()
```

```python
doge_df.reset_index(inplace=True)
doge_df.set_index('Date', inplace=True)
```

```python
doge_df['200_MA'] = doge_df['Close'].rolling('200D').mean()
doge_df['150_MA'] = doge_df['Close'].rolling('150D').mean()
doge_df['50_MA'] = doge_df['Close'].rolling('50D').mean()
doge_df['52W_High'] = doge_df['Close'].rolling('365D').max()
doge_df['52W_Low'] = doge_df['Close'].rolling('365D').min()
doge_df['Volatility']=(doge_df['High']-doge_df['Low'])/doge_df['Open']
doge_df.head(366)
```

```python
btc_df.reset_index(inplace=True)
btc_df.set_index('Date', inplace=True)
btc_df['200_MA'] = btc_df['Close'].rolling('200D').mean()
btc_df['150_MA'] = btc_df['Close'].rolling('150D').mean()
btc_df['50_MA'] = btc_df['Close'].rolling('50D').mean()
btc_df['52W_High'] = btc_df['Close'].rolling('365D').max()
btc_df['52W_Low'] = btc_df['Close'].rolling('365D').min()
btc_df['Volatility']=(btc_df['High']-btc_df['Low'])/btc_df['Open']
btc_df.head(366)
```

```python
eth_df.reset_index(inplace=True)
eth_df.set_index('Date', inplace=True)
eth_df['200_MA'] = eth_df['Close'].rolling('200D').mean()
eth_df['150_MA'] = eth_df['Close'].rolling('150D').mean()
eth_df['50_MA'] = eth_df['Close'].rolling('50D').mean()
eth_df['52W_High'] = eth_df['Close'].rolling('365D').max()
eth_df['52W_Low'] = eth_df['Close'].rolling('365D').min()
eth_df['Volatility']=(eth_df['High']-eth_df['Low'])/eth_df['Open']
eth_df.head(366)
```

```python
btc_df.to_csv('data/btc_data.csv')
eth_df.to_csv('data/eth_data.csv')
doge_df.to_csv('data/doge_data.csv')
```
