```python
import pandas as pd 
```

```python
df = pd.read_csv('data/btc_data.csv')
df['Pct_Chg'] = (df['Close']-df['Open'])/df['Open']
df
```

```python
import numpy as np
conditions = [
    (df['Pct_Chg'] >= .05), (df['Pct_Chg'] >= .01),
    (df['Pct_Chg'] >= -0.01), (df['Pct_Chg'] >= -.03)]
choices = ['Strong Buy', 'Buy', 'Hold','Weak Sell']
df['Category'] = np.select(conditions, choices, default='Liquidate')
#df = df.drop('Coin',axis=1)
#df = df.drop('Date',axis=1)
df
```

```python
X = df.drop('Category', axis=1)
Y = df['Category']
print(X.shape, Y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import TimeSeriesSplit
```

```python
def splitY(a, test_size=0.2):
    split_row = len(a) - int(test_size * len(a))
    y_train = a.iloc[:split_row]
    y_test = a.iloc[split_row:]
    return y_train, y_test
def train_test_splitX(x, test_size=0.2):
    split_row = len(x) - int(test_size * len(x))
    X_train = x.iloc[:split_row]
    X_test = x.iloc[split_row:]
    return X_train, X_test
X_train.shape
```

```python
X_train, X_test = train_test_splitX(X, test_size=0.2)
y_train, y_test = splitY(Y, test_size=0.2)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=13))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=5, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='mean_absolute_error',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=400,
    #shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:15])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:15])}")
```
